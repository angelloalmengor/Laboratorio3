Ejercicio 1
¿Cuál es el tamaño en bytes de un array de 1000 elementos de tipo booleano?

```python
import numpy as np

# Crear un array numpy de 1000 elementos de tipo booleano
arr_bool = np.array([True] * 1000)

# Obtener el tamaño en bytes del array
size_bytes = arr_bool.itemsize * arr_bool.size

# Imprimir el tamaño en bytes del array
print("El tamaño en bytes del array de 1000 elementos de tipo booleano es:", size_bytes)
```

Ejercicio 4

Dados los arrays a = [1, 4, 2, 7] y b = [1, 3, 2, 9],obtenga la media aritmética de la diferencia (a-b).

```python
# Definir los arrays a y b
a = np.array([1, 4, 2, 7])
b = np.array([1, 3, 2, 9])

# Calcular la diferencia entre los arrays a y b
diferencia = a - b

# Calcular la media aritmética de la diferencia
media_diferencia = np.mean(diferencia)

# Imprimir el resultado
print("La diferencia entre a y b es:", media_diferencia)
```

Ejercicio 5
Tengo valores de cordenadas (x, y) en las columnas del array

30.0 10.7
70.2 0.5
Agregue a este array las coordenadas (37.1, -3.6). Muestre en pantalla las dimensiones del nuevo array.

```python
# Definir el array inicial
array_inicial = np.array([[1.33, 4.5],
                          [30.0, 10.7],
                          [70.2, 0.5]])

# Coordenadas a agregar
coordenadas_nuevas = np.array([[37.1, -3.6]])

# Agregar las nuevas coordenadas al array existente
nuevo_array = np.vstack((array_inicial, coordenadas_nuevas))

# Mostrar las dimensiones del nuevo array
dimensiones_nuevo_array = nuevo_array.shape
print("Las dimensiones del nuevo array son:", dimensiones_nuevo_array)
```

Ejercicio 5: Obtenga la media, mediana, desviación estándar, valor máximo y valor mínimo de la población de los municipios de la provincia de Granada.

```python
import pandas as pd

# Filtrar los municipios de la provincia de Granada
municipios_granada = dataset[dataset['Provincia'] == 'Granada']

# Calcular las estadísticas de la población
media_poblacion = municipios_granada['Poblacion'].mean()
mediana_poblacion = municipios_granada['Poblacion'].median()
desviacion_estandar_poblacion = municipios_granada['Poblacion'].std()
valor_maximo_poblacion = municipios_granada['Poblacion'].max()
valor_minimo_poblacion = municipios_granada['Poblacion'].min()

# Imprimir los resultados
print("Media de la población en la provincia de Granada:", media_poblacion)
print("Mediana de la población en la provincia de Granada:", mediana_poblacion)
print("Desviación estándar de la población en la provincia de Granada:", desviacion_estandar_poblacion)
print("Valor máximo de la población en la provincia de Granada:", valor_maximo_poblacion)
print("Valor mínimo de la población en la provincia de Granada:", valor_minimo_poblacion)
```

Ejercicio 7: Seleccione al azar cincuenta municipios diferentes de entre los diponibles en el archivo. Asegúrese de que no se repitan. ¿Luego calcule el promedio de la población y la desviación estándar de esto 50 municipios?.

```python
# Seleccionar 50 municipios al azar sin reemplazo
muestras_aleatorias = dataset.sample(n=50, replace=False)

# Calcular el promedio y la desviación estándar de la población de las muestras seleccionadas
promedio_poblacion = muestras_aleatorias['Poblacion'].mean()
desviacion_estandar_poblacion = muestras_aleatorias['Poblacion'].std()

# Imprimir los resultados
print("Promedio de la población de los 50 municipios seleccionados al azar:", promedio_poblacion)
print("Desviación estándar de la población de los 50 municipios seleccionados al azar:", desviacion_estandar_poblacion)
```

Ejercicio 8: Determine los nombres de los Municipios más cercano y más lejano a Madrid. Para ello debe calcular la distancia en todos ellos y Madrid. No considere a Madrid en el análisis ya que la distancia sería cero

```python
madrid_data = dataset[dataset["Municipio"] == "Madrid"]
madrid_coordsX = (madrid_data["X"].values[0])
madrid_coordsY = (madrid_data["Y"].values[0])
```

```python
madrid_coords = (madrid_coordsY, madrid_coordsX)

dataset["Distancia_a_Madrid"] = ((dataset["X"] - madrid_coords[0])**2 + (dataset["Y"] - madrid_coords[1])**2)**0.5

# Encontrar el municipio más cercano y el más lejano a Madrid
municipio_mas_cercano = dataset.loc[dataset["Distancia_a_Madrid"].idxmin()]["Municipio"]
municipio_mas_lejano = dataset.loc[dataset["Distancia_a_Madrid"].idxmax()]["Municipio"]

print("Municipio más cercano a Madrid:", municipio_mas_cercano)
print("Municipio más lejano a Madrid:", municipio_mas_lejano)
```
